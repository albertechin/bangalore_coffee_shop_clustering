```python
def geocode(postalcode):
    API_KEY = 'dc42f8f1757c17'
    url = 'https://us1.locationiq.com/v1/search.php'
    params = {'key': API_KEY, 'format':'json', 'postalcode':postalcode}
    r = requests.get(url, params=params)
    result = r.json()
    time.sleep(1) # Only 60 requests are allowed per minute so delaying each request by sec 
    print((result[0]['lat'], result[0]['lon']))
    return (result[0]['lat'], result[0]['lon'])
```

```python
# bng_df['Latitude'] = bng_df['Postcode'].apply(lambda x: geocode(x)[0])
```

```python
lat = []
lng = []

def lat_lng_list_gen(postalcode):
    lat_lng = geocode(postalcode)
    lat.append(lat_lng[0])
    lng.append(lat_lng[1])
```

```python
bng_df['pincode'].apply(lambda x: lat_lng_list_gen(x))
```

```python
bng_df['Latitude'] = lat
```

```python
bng_df['Longtude'] = lng
```
